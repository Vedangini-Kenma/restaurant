```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path='/content/drive/MyDrive/restaurants  - Sheet1.csv'
df=pd.read_csv(path)
df
```

**Identifing** **Outliers** **with** **IQR**

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.scatter( df['views'],df['stars'])
ax.set_xlabel('view')
ax.set_ylabel('star')
print("variation in views though stars are same")
plt.show()
```

**comparing** **after** **trimming**

```python
new_df = df[(df['views'] >= 3170 ) | (df['views'] <= 1292)]
#cf = cf[cf]
new_df
```

**Plotting** **Graph** **to** **underatand** 

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['views'])
plt.subplot(2,2,2)
sns.boxplot(df['views'])
plt.subplot(2,2,3)
sns.distplot(new_df['views'])
plt.subplot(2,2,4)
sns.boxplot(new_df['views'])
plt.show()
```
